```python
import nltk  #natural language toolkit
import re
from nltk.corpus import stopwords
```

```python
#downloading the stopwords and define
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
# Function to clean the text
def clean_text(text):

    text = text.lower()

    # Remove punctuation and non-alphabetic characters
    text = re.sub(r'[^a-z\s]', '', text)


    words = text.split()

    # Remove stopwords
    words = [word for word in words if word not in stop_words]

    return ' '.join(words)

# Apply cleaning function to the TITLE and ABSTRACT columns
df['cleaned_title'] = df['TITLE'].apply(clean_text)
df['cleaned_abstract'] = df['ABSTRACT'].apply(clean_text)

# Combine cleaned TITLE and ABSTRACT into a single text column
df['cleaned_text'] = df['cleaned_title'] + ' ' + df['cleaned_abstract']

# Display the cleaned data
df[['TITLE', 'ABSTRACT', 'cleaned_text']].head()
```

 Tokenization & Lemmatization:
We can perform tokenization (splitting into words) and lemmatization (reducing words to their base form).

```python
import spacy

# Load the small English model
nlp = spacy.load('en_core_web_sm')
# Lemmatization form example= running convert into run
# Lemmatization function using batch processing
def lemmatize_texts(texts):
    # Process the texts in a batch
    docs = nlp.pipe(texts, disable=['parser', 'ner'], batch_size=50)
    # Extract lemmatized words
    lemmatized_texts = []
    for doc in docs:
        lemmatized_texts.append(' '.join([token.lemma_ for token in doc if token.is_alpha]))
    return lemmatized_texts

# Apply lemmatization to the cleaned text using batch processing
df['lemmatized_text'] = lemmatize_texts(df['cleaned_text'].tolist())  # Convert to list for batch processing

# Display the first few lemmatized texts
df[['cleaned_text', 'lemmatized_text']].head(100)
```

```python
from sklearn.feature_extraction.text import CountVectorizer   # used for converting a collection of text documents to a matrix of token counts


vectorizer = CountVectorizer()

# Create the Document-Term Matrix
dtm = vectorizer.fit_transform(df['lemmatized_text'])    # fit=assigns an index to each unique word and transform=creates a sparse matrix representation of the documents

# Get the feature names (terms)
terms = vectorizer.get_feature_names_out()

# Convert to DataFrame for better visualization
dtm_df = pd.DataFrame(dtm.toarray(), columns=terms)

# Display the Document-Term Matrix
dtm_df.head()
```

 Train the LDA(Latent Dirichlet Allocation) Model

 lda=discover hidden topics in a collection of documents





```python
import gensim  # gensim =used for analysis document similarity
from gensim import corpora

dictionary = corpora.Dictionary([text.split() for text in df['lemmatized_text']])

corpus = [dictionary.doc2bow(text.split()) for text in df['lemmatized_text']]

num_topics = 5

# Train the LDA model
lda_model = gensim.models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=15)


topics = lda_model.print_topics(num_words=4)
for topic in topics:
    print(topic)
```
